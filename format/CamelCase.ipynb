```python
import json
```

```python
file_path = "../json/qualities.json"
```

```python
with open(file_path, 'r') as file:
    f = json.load(file)
    for k in f:
        for i in range(len(f[k])):
            f[k][i] = f[k][i].title()

with open(file_path, 'w') as file:
    json.dump(f, file, indent=4)
```
